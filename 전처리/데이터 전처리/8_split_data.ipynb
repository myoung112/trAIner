```python
# ver1

import os
import shutil
import random
import math

exer = ['BicycleCrunch',
        'BurpeeTest', 
        'CrossLunge', 
        'Crunch', 
        'Flank', 
        'Goodmorning', 
        'HipThrust', 
        'KneePushup', 
        'LyingLegRaise', 
        'Pushup', 
        'ScissorCross', 
        'SideLunge',
        'StandingKneeUp', 
        'StandingSideCrunch', 
        'StepBackwardDynamicLunge', 
        'StepForwardDynamicLunge'
       ]

for exercise in exer:
    # 원본 폴더와 대상 폴더 경로 설정
    source_folder = f'E:\\AI\\dataset_only_skeleton\\face\\{exercise}\\training'
    valid_folder = f'E:\\AI\\dataset_only_skeleton\\face\\{exercise}\\validation'
    test_folder = f'E:\\AI\\dataset_only_skeleton\\face\\{exercise}\\test'
    
    path = []
    
    for item in os.listdir(source_folder):
        path.append(item)
    
    # 원본 폴더에서 .jpg 파일만 리스트로 가져오기
    for exe in path:
        source = os.path.join(source_folder, exe)
        valid = os.path.join(valid_folder, exe)
        test = os.path.join(test_folder, exe)
        
        # 검증(validation) 및 테스트(test) 폴더가 없으면 생성
        os.makedirs(valid, exist_ok=True)
        os.makedirs(test, exist_ok=True)
        
        files = [file for file in os.listdir(source) if file.endswith('.jpg')]
        
        total_files_count = len(files)
        sample_size = math.ceil(total_files_count * 0.3)  # 전체 파일의 30% 계산
        valid_test_size = math.ceil(sample_size * 0.5)  # 선택된 파일의 50% 계산 (전체의 15%)
    
        if total_files_count < sample_size:
            print("Error: 원본 폴더에 필요한 만큼의 .jpg 파일이 없습니다.")
        else:
            # 파일을 무작위로 선택하여 30% 리스트 생성
            selected_files = random.sample(files, sample_size)
            
            # 검증(validation) 폴더로 이동할 파일 선택
            valid_files = selected_files[:valid_test_size]
            
            # 테스트(test) 폴더로 이동할 파일 선택
            test_files = selected_files[valid_test_size:2*valid_test_size]
            
            # 선택된 파일을 검증(validation) 폴더로 이동
            for file in valid_files:
                shutil.move(os.path.join(source, file), os.path.join(valid, file))
            
            # 선택된 파일을 테스트(test) 폴더로 이동
            for file in test_files:
                shutil.move(os.path.join(source, file), os.path.join(test, file))
            
            print(f"{len(valid_files)}개의 파일이 검증(validation) 폴더로, {len(test_files)}개의 파일이 테스트(test) 폴더로 이동되었습니다.")
```
